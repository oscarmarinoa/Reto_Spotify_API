### Tarea Parte 1

```python
# Import the necessary libraries.
import json
import pandas as pd
import numpy as np
```

```python
f = open(r'taylor_swift_spotify.json')
```

```python
data = json.load(f)
```

```python
dict_taylor = {'disc_number': [],
               'duration_ms': [],
               'explicit': [],
               'track_number': [],
               'track_popularity': [],
               'track_id': [], 
               'track_name': [],
               'audio_features.danceability': [], 
               'audio_features.energy': [], 
               'audio_features.key': [], 
               'audio_features.loudness': [], 
               'audio_features.mode': [], 
               'audio_features.speechiness': [], 
               'audio_features.acousticness': [], 
               'audio_features.instrumentalness': [], 
               'audio_features.liveness': [], 
               'audio_features.valence': [], 
               'audio_features.tempo': [], 
               'audio_features.id': [], 
               'audio_features.time_signature': [],
               'artist_id': [],
               'artist_name' : [],
               'artist_popularity' : [],
               'album_id': [], 
               'album_name': [], 
               'album_release_date': [], 
               'album_total_tracks': []}
```

```python
for key in data.keys(): #artist_id
    
    if      key == 'artist_id': value_1 = data[key]
    elif    key == 'artist_name': value_2 = data[key]
    elif    key == 'artist_popularity': value_3 = data[key]
    
    if type(data[key]) == dict or type(data[key]) == list:
        
        for no_dict in range(len(data[key])): # diccionarios en albums
            
            
            for sub_dict in data[key][no_dict].keys(): # #album id, album_name, 
                
                if      sub_dict == 'album_id': value_4 = data[key][no_dict][sub_dict]
                elif    sub_dict == 'album_name': value_5 = data[key][no_dict][sub_dict]
                elif    sub_dict == 'album_release_date': value_6 = data[key][no_dict][sub_dict]
                elif    sub_dict == 'album_total_tracks': value_7 = data[key][no_dict][sub_dict]
                
                if type(data[key][no_dict][sub_dict]) == dict or type(data[key][no_dict][sub_dict]) == list: #tracks
                    
                    for no_dict_2 in range(len(data[key][no_dict][sub_dict])): # for dictionaries in tracks
                        
                        dict_taylor['artist_id'].append(value_1)
                        dict_taylor['artist_name'].append(value_2)  
                        dict_taylor['artist_popularity'].append(value_3)
                        dict_taylor['album_id'].append(value_4)
                        dict_taylor['album_name'].append(value_5)  
                        dict_taylor['album_release_date'].append(value_6)  
                        dict_taylor['album_total_tracks'].append(value_7)  
                        
                        for sub_sub_dict in data[key][no_dict][sub_dict][no_dict_2].keys():                                                        
                            
                            if type(data[key][no_dict][sub_dict][no_dict_2][sub_sub_dict]) == dict or type(data[key][no_dict][sub_dict][no_dict_2][sub_sub_dict]) == list:
                                
                                for sub_sub_sub_dict in  data[key][no_dict][sub_dict][no_dict_2][sub_sub_dict].keys():                         
                                    
                                    auxiliar = 'audio_features.' + sub_sub_sub_dict
                                        
                                    dict_taylor[auxiliar].append(data[key][no_dict][sub_dict][no_dict_2][sub_sub_dict][sub_sub_sub_dict])

                            else:
                                dict_taylor[sub_sub_dict].append(data[key][no_dict][sub_dict][no_dict_2][sub_sub_dict])           
                          
```

```python
for i in dict_taylor.keys():
    print(len(dict_taylor[i]))
```

```python
df = pd.DataFrame(dict_taylor)
df.to_csv('dataset2.csv', index=False)
```
